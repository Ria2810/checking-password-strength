```python
# Define the expected number of columns in your CSV file
expected_number_of_columns = 2  # Replace with the actual number of columns

# Define the file paths
input_csv_file = 'C:/Users/riach/Projects/Checking Password Strength/data.csv'
output_csv_file = 'cleaned_data.csv'

# Read and clean the CSV file
with open(input_csv_file, 'r', encoding='utf-8') as file:
    lines = file.readlines()
    cleaned_lines = [line for line in lines if len(line.split(',')) == expected_number_of_columns]

# Write the cleaned lines back to a new file or overwrite the original file
with open(output_csv_file, 'w', encoding='utf-8') as file:
    file.writelines(cleaned_lines)

# Now, you can read the cleaned CSV file using pandas
data = pd.read_csv(output_csv_file)
```

```python
data.dropna(inplace=True)
```

```python
passwords_tuple=np.array(data)
passwords_tuple
```

```python
import random
random.shuffle(passwords_tuple) #shuffling randomly for robustness
```

```python
y=[labels[1] for labels in passwords_tuple]
y
```

```python
X=[labels[0] for labels in passwords_tuple]
```

```python
def word_divide_char(inputs):
    characters=[]
    for i in inputs:
        characters.append(i)
    return characters
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(tokenizer=word_divide_char)
X=vectorizer.fit_transform(X)
```

```python
X.shape
```

```python
vectorizer.vocabulary_
```

```python
data.iloc[0,0]
```

```python
feature_names = vectorizer.get_feature_names()
 
#get tfidf vector for first document
first_document_vector=X[0]
 
#print the scores
df = pd.DataFrame(first_document_vector.T.todense(), index=feature_names, columns=["tfidf"])
df.sort_values(by=["tfidf"],ascending=False)
```

```python
## Logistics Regression

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)  #splitting
```

```python
log_class=LogisticRegression(penalty='l2',multi_class='ovr')
log_class.fit(X_train,y_train)
```

```python
print(log_class.score(X_test,y_test))
```

```python
## Multinomial

clf = LogisticRegression(random_state=0, multi_class='multinomial', solver='newton-cg')
clf.fit(X_train, y_train) #training
print(clf.score(X_test, y_test))
```

```python
X_predict=np.array(["@7abbtw^&"])
X_predict=vectorizer.transform(X_predict)
y_pred=log_class.predict(X_predict)
print(y_pred)
```

```python
import pickle
# open a file, where you ant to store the data
file = open('LogisticRegression_classifier_model.pkl', 'wb')

# dump information to that file
pickle.dump(log_class, file)
```
